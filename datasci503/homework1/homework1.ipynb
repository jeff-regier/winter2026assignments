**Problem 3:** Bias-Variance Tradeoff

You have a dataset. You are considering a collection of different methods that you could apply to this dataset in order to estimate the regression function. The methods range from very inflexible (only capable of representing a small class of true regression functions) to very flexible (capable of fitting a very large class of true regression functions).

Answer True or False for each statement, and provide a brief explanation.

**(a)** Typically, the bias of your estimate will be lower with more flexible methods.

> BEGIN SOLUTION

True. Flexible methods are more capable of being able to represent our regression function which will directly result in the final model having lower bias.

> END SOLUTION

**(b)** Typically, the variance of your estimate will be lower with more flexible methods.

> BEGIN SOLUTION

False. Flexible methods are able to more accurately represent the regression function due to their ability to take into account more features and data, but at the same time they also end up picking up noise and irregularities which can result in them having a higher variance than inflexible methods which are highly selective and restrictive.

> END SOLUTION

**(c)** Typically, irreducible error will be lower with more flexible methods.

> BEGIN SOLUTION

False. Irreducible error is the random error that exists in our data. No matter which type of model we use or how flexible that model is, the irreducible error will remain the same. Therefore, both the flexible and the highly inflexible model should have the same irreducible error.

> END SOLUTION

**Problem 4:** Asymptotic Properties

Consider gathering a dataset and using it to estimate the regression function. Assume the data-generating distribution is well behaved (e.g., assume that the true regression function is continuous) and typical of what we would find in real-world settings. Answer True or False.

**(a)** As the size of your training dataset tends to infinity (for a fixed number of neighbors, $K$), the bias of the KNN regression estimate will tend to 0.

> BEGIN SOLUTION

True. As the training dataset grows to infinity, for a fixed $K$, the $K$ nearest neighbors will become arbitrarily close to the query point. Since we assume the regression function is continuous, the true function values at these nearby points will converge to the true value at the query point, causing the bias to tend to zero.

> END SOLUTION

**(b)** As the size of your training dataset tends to infinity (for a fixed number of neighbors, $K$), the variance of the KNN regression estimate will tend to 0.

> BEGIN SOLUTION

False. The variance of the KNN estimate depends on the variance of the response values at the $K$ nearest neighbors, not on how many training points exist overall. Since $K$ is fixed, we always average over exactly $K$ neighbors. As $n \to \infty$, these neighbors get closer together spatially, but the noise in their responses remains, so the variance does not tend to zero.

> END SOLUTION

**(c)** As the size of your training dataset tends to infinity, the bias of a least squares linear regression estimate will tend to zero.

> BEGIN SOLUTION

False. Linear regression is constrained to fit a linear function. In real-world settings, the true regression function is almost never exactly linear. Therefore, regardless of how much data we have, the linear model will be misspecified and unable to capture the true nonlinear relationship. This model misspecification causes bias that persists even as $n \to \infty$.

> END SOLUTION

**(d)** As the size of your training dataset tends to infinity, the variance of a least squares linear regression estimate will tend to 0.

> BEGIN SOLUTION

True. The variance of the least squares estimator is proportional to $1/n$. As the sample size $n$ tends to infinity, the coefficient estimates become increasingly stable because we are averaging over more data points. Therefore, the variance of the linear regression estimate tends to zero as $n \to \infty$.

> END SOLUTION

**Problem 5e:** For Loops

Write a function called `generate_hello_messages` that uses a for loop to return a list containing the following strings:
```
["Hello world 1.", "Hello world 2.", "Hello world 4.", "Hello world 8.", "Hello world 16."]
```

Then call your function to print each message on its own line.

**Hint:** Notice the pattern in the numbers: they are powers of 2 ($2^0, 2^1, 2^2, 2^3, 2^4$).

````python
# BEGIN SOLUTION
def generate_hello_messages():
    """Generate a list of 'Hello world' messages with powers of 2."""
    messages = []
    for i in range(5):
        messages.append(f"Hello world {2**i}.")  # noqa: PERF401
    return messages


# END SOLUTION

# Print each message
for message in generate_hello_messages():
    print(message)
````

````python
# Test assertions
messages = generate_hello_messages()
expected = [
    "Hello world 1.",
    "Hello world 2.",
    "Hello world 4.",
    "Hello world 8.",
    "Hello world 16.",
]
assert messages == expected, f"Expected {expected}, got {messages}"
print("All tests passed!")

# BEGIN HIDDEN TESTS
assert len(messages) == 5, "Should have exactly 5 messages"
assert messages[0] == "Hello world 1.", "First message should be 'Hello world 1.'"
assert messages[-1] == "Hello world 16.", "Last message should be 'Hello world 16.'"
assert all("Hello world" in m for m in messages), "All messages should contain 'Hello world'"
# END HIDDEN TESTS
````
